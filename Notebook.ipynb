## Visualization & Explanatory Analysis

### Pertanyaan 1:
Stasiun apa yang memiliki temperatur paling tinggi dan paling rendah?

```python
sum_temp_df = df.groupby("station").TEMP.mean().sort_values(ascending=False).reset_index()
sum_temp_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="TEMP", y="station", data=sum_temp_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Temperatur Paling Tinggi", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="TEMP", y="station", data=sum_temp_df.sort_values(by="TEMP", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Temperatur Paling Rendah", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Temperatur Paling Tinggi dan Rendah berdasarkan Stasiun", fontsize=20)
plt.show()
```

### Pertanyaan 2:
Bagaimana partikel udara dengan ukuran 10 yang ada terhadap arah angin?

```python
bywd_df = df.groupby(by="wd").PM10.nunique().reset_index()
bywd_df.rename(columns={
    "PM10": "count"
}, inplace=True)

plt.figure(figsize=(10, 5))

sns.barplot(
    y="count",
    x="wd",
    data=bywd_df.sort_values(by="count", ascending=False),
    palette=colors
)
plt.title("Partikel Udara 10 terhadap Arah Angin", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```
